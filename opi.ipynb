```python
import pandas as panda
import seaborn as scs
scs.set(color_codes=True)
```

```python
weather = panda.read_csv('weather.csv')
```

Now I will **display** The content of the *.csv* with a good-looking table.

```python
weather.head()
```

This is a way to display the info about the *.csv* file:

```python
weather.info()
```

```python
scs.barplot(x=weather['temperature'], y=weather['humidity'])
```

```python
scs.distplot(weather['humidity'])
```

```python
scs.jointplot(x=weather['temperature'], y=weather['humidity'], kind='hex')
```

```python
scs.jointplot(x=weather['temperature'], y=weather['humidity'], kind='kde')
```

```python
scs.pairplot(weather[['humidity', 'temperature']])
```

```python
scs.stripplot(x=weather['temperature'], y=weather['humidity'])
```
